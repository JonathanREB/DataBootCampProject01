```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_payday = pd.read_csv('input/sales1.csv')

df_payday.head()
```

```python
list(df_payday)
```

```python
info_payday = ['Order Date','Postcode','Status', 'Payment', 'Item', 'Total net', 'Item Type', 'Ship Date', 'State', 'Delivery Date']
df_payday = df_payday[info_payday]
df_payday.head()
```

```python
date_payday_df = df_payday.groupby('Order Date')
date_payday_df.head()
```

```python
Shipped = df_payday['Status'] == 'Shipped'
print(Shipped.head())
```

```python
shipped_payday_df = df_payday[Shipped]
print(shipped_payday_df.head())
```

```python
shipped_payday_df.Payment.unique()
```

```python
consumer = shipped_payday_df['Payment'] == 'Wholesale'
print(consumer.head())
```

```python
#aqui estoy excluyendo las ventas de mayoreo pero no puedo hacerlo con solo un valor, entonces tambien exclui el valor mercado
mask = shipped_payday_df['Payment'].isin(['Wholesale', 'Mercado'])
shipped_payday_df[~mask]
```

```python
consumer_shipped_payday_df = shipped_payday_df[~mask]
consumer_shipped_payday_df.head()
```

```python
shipped_date1 = consumer_shipped_payday_df['Order Date'].str.split('/', 1, expand = True)
shipped_date1.head()
```

```python
shipped_date1.rename(  columns={
    0 : 'month',
    1 : 'day_year'
  },
  inplace=True)
shipped_date1.head()
```

```python
shipped_date2 = shipped_date1['day_year'].str.split('/', 1, expand = True)
shipped_date2.head()
```

```python
shipped_date2.rename(  columns={
    0 : 'day',
    1 : 'year'
  },
  inplace=True)
shipped_date2.head()
```

```python
shipped_date3 = shipped_date2.join(shipped_date1, how = 'outer')
shipped_date3.head()
```

```python
shipped_date4 = shipped_date3.drop('day_year', axis = 1)
shipped_date4.head()
```

```python
enriched_shipped_payday_df = consumer_shipped_payday_df.join(shipped_date4,how ='outer')
enriched_shipped_payday_df.head()
```

```python
enriched_shipped_payday_df['sales'] = 1
enriched_shipped_payday_df.head()
```

```python
enriched_shipped_payday_df['Payment'].unique()
```

```python
shipped_date_df = enriched_shipped_payday_df.groupby('day')
shipped_date_df.head()
```

```python
grouped_day_df = enriched_shipped_payday_df.groupby('day')
grouped_day_df.aggregate(np.sum)
```

```python
#aqui quisiera graficar dia vs sales pero no puedo
grouped_day_df.plot(kind='line',x='day',y='sales')
plt.show()
```
